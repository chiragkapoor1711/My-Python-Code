```python
num=6
mum=5
for i in range(1,num+1):
    for s in range(num-i):
        print(" ",end=" ")
    for str in range(i):
        print("*",end=" ")
    print()
for i in range(mum,0,-1):
    for s in range(num-i):
        print(" ",end=" ")
    for str in range(i):
        print("*",end=" ")
    print()
```

**Random randint game**

```python
import random
while True:
    secret = random.randint(1,3)
    guess = int(input())
    if guess == secret:
        print("you won")
        break
        
```
