**lambda**

```python
def add(a,b):
    return a+b
add(10,5)
```

```python
add = lambda a,b : a + b
add(10,12)
```

```python
is_even = lambda a : a % 2 ==0
is_even(8)
```

```python
greater = lambda a,b : a if a > b else b
greater(5,7)
```

**map**

```python
def sqr(x):
    return x * x
sqr(4)
```

```python
lst = [1,2,3,4]
res= map(lambda a : a * a,lst)
print(list(res))
```

**filter**

```python
lst=[1,2,3,4]
res=filter(lambda a: a % 2 == 0, lst)
print(list(res)
```
