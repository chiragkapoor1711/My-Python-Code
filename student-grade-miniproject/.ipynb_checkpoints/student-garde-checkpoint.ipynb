```python
students_grade={}
file_name="studnets.txt"

def load_from_file():
    try:
        with open(file_name,"r") as file:
            for line in file:
                line=line.strip()
                if line:
                    name ,grade = line.split("-")
                    students_grade[name]=grade
    except FileNotFoundError:
        pass

def save_to_file():
    with open(file_name,"w") as file:
        for name,grade in students_grade.items():
            file.write(f"{name}-{grade}\n")
    
        
        
def add_student(name,grade):
    students_grade[name]=grade
    save_to_file()
    print(f"added {name} with garde {grade}")
def update_student(name,grade):
    students_grade[name]=grade
    save_to_file()
    if name in students_grade:
        print(f"{name} with marks are updated {grade}")
    else:
        print(f"{name} student not found")
def delete_student(name):
    if name in students_grade:
        del students_grade[name]
        save_to_file()
        print(f"{name} data is deleted")
    else:
        print(f"{name} not found")
def view_student():
    if students_grade:
        for name , grade in students_grade.items():
            print(f"{name} : {grade}")
    else:
        print("no data found")
def main():
    load_from_file()
    while True:
        print("Studnets grade managemanagement system")
        print("1. Add student")
        print("2. Update student")
        print("3. Delete student")
        print("4. View student")
        print("5. Exit")
        choice = input("enter your choice")
        if choice == '1':
            name=input("enter student name")
            grade=input("enter grade of student")
            add_student(name,grade)
        elif choice =='2':
            name=input("enter studnet name")
            grade=input("enter updated grade")
            update_student(name,grade)
        elif choice =='3':
            name=input("enter delete student name")
            delete_student(name)
        elif choice =='4':
            view_student()
        elif choice =='5':
            print("program end")
            break
        else:
            print("invalid choice")
            
main()
  
```
