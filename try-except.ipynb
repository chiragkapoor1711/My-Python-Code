```python
a = int(input('num1'))
b =int(input('num2'))
print(a/b)
```

```python
try:
    a=int(input('num1'))
    b=int(input('num2'))
    print(a/b)
except (ZeroDivisionError):
    print('number is divisble by 0')
except (ValueError):
    print('enter number only')
finally:
    print('program run successfully')
```

```python
try:
    a=int(input('num1'))
except (ValueError):
    print('taking default number')
    a=0
try:
    b=int(input('num2'))
except (ValueError):
    print('takig default value')
    b=0
try:
    print(a/b)
except (ZeroDivisionError):
    print('cannot be divideed by zero')
    print('mul',a*b)
    print('sum',a+b)
```
